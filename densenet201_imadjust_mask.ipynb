```python
import os
import time  # Zaman ölçümü için
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, precision_score, recall_score, f1_score

import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import DenseNet201
from tensorflow.keras import layers, models
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam
from google.colab import drive

# Google Drive'ı bağla
drive.mount('/content/drive')

# Veri yolları
train_dir = '/content/drive/MyDrive/imadjust/train'
test_dir = '/content/drive/MyDrive/imadjust/test'

# Veri toplama fonksiyonu
def gather_paths_and_labels(directory):
    data = []
    for label, class_name in enumerate(['Normal', 'Scol']):
        class_dir = os.path.join(directory, class_name)
        for file in os.listdir(class_dir):
            data.append((os.path.join(class_dir, file), label))
    return np.array(data)

# Verileri yükleme
train_data = gather_paths_and_labels(train_dir)
test_data = gather_paths_and_labels(test_dir)

# Class Weights
class_weights_dict = {0: 1.0, 1: 0.7}
print(f"\nUpdated Class Weights: {class_weights_dict}")

# Veri artırma (Data Augmentation)
train_datagen = ImageDataGenerator(
    rescale=1./255, rotation_range=15, width_shift_range=0.2,
    height_shift_range=0.2, zoom_range=[0.8, 1.2], horizontal_flip=True
)
val_test_datagen = ImageDataGenerator(rescale=1./255)

# DenseNet201 Modeli
def create_densenet201_model():
    base_model = DenseNet201(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

    for layer in base_model.layers[:-10]:  # Son 10 katman dışında tümünü dondur
        layer.trainable = False
    for layer in base_model.layers[-10:]:
        layer.trainable = True

    x = layers.GlobalAveragePooling2D()(base_model.output)
    x = layers.Dense(256, activation='relu')(x)
    x = layers.Dropout(0.5)(x)
    output = layers.Dense(1, activation='sigmoid')(x)

    model = models.Model(inputs=base_model.input, outputs=output)
    model.compile(optimizer=Adam(learning_rate=1e-4),
                  loss="binary_crossentropy",
                  metrics=["accuracy"])
    return model

# Cross-Validation Süresi ve Train Accuracy Ölçümü
def run_cross_validation(data, n_splits=5, epochs=50, batch_size=32):
    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)

    accuracy_list, fold_times = [], []

    total_start_time = time.time()  # Cross-validation süresini başlat

    for fold, (train_idx, val_idx) in enumerate(skf.split(data[:, 0], data[:, 1])):
        print(f"\n=== Fold {fold + 1}/{n_splits} ===")

        fold_start_time = time.time()

        train_files, val_files = data[train_idx], data[val_idx]
        train_df = pd.DataFrame({'filename': train_files[:, 0], 'class': train_files[:, 1]})
        val_df = pd.DataFrame({'filename': val_files[:, 0], 'class': val_files[:, 1]})

        train_gen = train_datagen.flow_from_dataframe(
            train_df, x_col='filename', y_col='class',
            target_size=(224, 224), batch_size=batch_size, class_mode='binary', shuffle=True
        )

        val_gen = val_test_datagen.flow_from_dataframe(
            val_df, x_col='filename', y_col='class',
            target_size=(224, 224), batch_size=batch_size, class_mode='binary', shuffle=False
        )

        model = create_densenet201_model()
        early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

        model.fit(train_gen, validation_data=val_gen, epochs=epochs, callbacks=[early_stopping], class_weight=class_weights_dict, verbose=1)

        fold_end_time = time.time()
        fold_time = fold_end_time - fold_start_time
        fold_times.append(fold_time)

        print(f"Fold {fold + 1} eğitim süresi: {fold_time:.2f} saniye")

        val_labels = val_files[:, 1].astype(int)
        val_probs = model.predict(val_gen).ravel()
        val_preds = (val_probs > 0.5).astype(int)
        accuracy = accuracy_score(val_labels, val_preds)

        accuracy_list.append(accuracy)

    total_end_time = time.time()
    total_cv_time = total_end_time - total_start_time

    mean_train_acc = np.mean(accuracy_list)
    std_train_acc = np.std(accuracy_list)

    print(f"\n⏳ Toplam Cross-Validation Süresi: {total_cv_time:.2f} saniye")
    print(f"Train Accuracy: {mean_train_acc:.4f} ± {std_train_acc:.4f}")

    return total_cv_time, mean_train_acc, std_train_acc

# Final Model Eğitimi ve Test Değerlendirme
def evaluate_final_model(model, test_data):
    test_df = pd.DataFrame({'filename': test_data[:, 0], 'class': test_data[:, 1]})
    test_gen = val_test_datagen.flow_from_dataframe(
        test_df, x_col='filename', y_col='class',
        target_size=(224, 224), batch_size=32, class_mode='binary', shuffle=False
    )

    test_labels = test_data[:, 1].astype(int)
    test_probs = model.predict(test_gen).ravel()
    test_preds = (test_probs > 0.5).astype(int)

    test_accuracy = accuracy_score(test_labels, test_preds)
    test_f1 = f1_score(test_labels, test_preds)
    test_precision = precision_score(test_labels, test_preds)
    test_recall = recall_score(test_labels, test_preds)


    cm = confusion_matrix(test_labels, test_preds)

    print("\n=== Final Model Test Sonuçları ===")
    print(f"Test Accuracy: {test_accuracy:.4f}")
    print(f"Test F1 Score: {test_f1:.4f}")
    print(f"Test Precision: {test_precision:.4f}")
    print(f"Test Recall: {test_recall:.4f}")

    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=["Normal", "Scol"], yticklabels=["Normal", "Scol"])
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.title("Confusion Matrix - Final Model")
    plt.show()

    return test_accuracy, test_f1, test_precision, test_recall

# Model Eğitimi
cv_time, mean_train_acc, std_train_acc = run_cross_validation(train_data, n_splits=5, epochs=50, batch_size=32)
final_model = create_densenet201_model()
final_model.fit(train_datagen.flow_from_directory(train_dir, target_size=(224, 224), batch_size=32, class_mode='binary'), epochs=50)
test_acc, test_f1, test_precision, test_recall = evaluate_final_model(final_model, test_data)

# Sonuçları Yazdırma
print("\n=== Sonuçlar ===")
print(f"Toplam Hesaplama Süresi: {cv_time:.2f} saniye")
print(f"Train Accuracy: {mean_train_acc:.4f} ± {std_train_acc:.4f}")
print(f"Test Accuracy: {test_acc:.4f}")
print(f"Test F1 Score: {test_f1:.4f}")
print(f"Test Precision: {test_precision:.4f}")
print(f"Test Recall: {test_recall:.4f}")
```
